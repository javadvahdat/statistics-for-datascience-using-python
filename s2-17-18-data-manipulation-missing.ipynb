 - **Adding rows to an existing Pandas DataFrame**

```python
dict = {'Name':['Martha', 'Tim', 'Rob', 'Georgia'],
        'Maths':[87, 91, 97, 95],
        'Science':[83, 99, 84, 76]
       }
  
df1 = pd.DataFrame(dict)
display(df1)
```

```python
len(df1.index)
```

```python
df1.loc[len(df1.index)] = ['Amy', 89, 93] 
  
display(df1)
```

```python
df2 = {'Name': 'Amy', 'Maths': 89, 'Science': 93}
df1 = df1.append(df2, ignore_index = True)
df1
```

```python
df3 = pd.concat([df1, df1], axis=1, ignore_index=True)
df3
```

**Dealing with missing data** 

```python
misdata = pd.read_csv("misdata.csv")
misdata
```

we note that Pandas automatically assigns `NaN` if the value for a particular column is an empty string `''` `NA` or `NaN`.

```python
misdata.isna().sum()
```

```python
missing_values = ["n.a.","NA","n/a", "na", 0]
misdata = pd.read_csv('misdata.csv', na_values = missing_values)
misdata
```

```python
misdata.isna().sum()
```

```python
misdata.dropna()  #remove rows with NaN
```

```python
misdata.dropna(axis=1)  #remove columns with NaN
```

```python
misdata.dropna(axis=0, how="all")  #all: if all values are NA, drop that label
```

**Filling out Missing Values**

```python
misdata["Salary"].median()
```

```python
misdata["Salary"].fillna(misdata["Salary"].median(), inplace=True) #also mean,mode
misdata
```

```python
misdata["Bonus %"].fillna(0, inplace=True) #Constant number
misdata
```

```python
missing_values = ["n.a.","NA","n/a", "na", 0]
misdata = pd.read_csv('misdata.csv', na_values = missing_values)
misdata
```

```python
misdata['Salary'].fillna(method='ffill', inplace=True) #Forward Fill
misdata
```

```python
misdata['Bonus %'].fillna(method='bfill', inplace=True) #Backward  Fill
misdata
```

```python
missing_values = ["n.a.","NA","n/a", "na", 0]
misdata = pd.read_csv('misdata.csv', na_values = missing_values)
```

```python
misdata['Salary'].interpolate(method='polynomial', order=2, inplace=True)
misdata
```

```python
missing_values = ["n.a.","NA","n/a", "na", 0]
misdata = pd.read_csv('misdata.csv', na_values = missing_values)
misdata.head()
```
